```python
# 5.Decide input and output
x = df.drop('outcome', axis=1) # Input / Independent F
y= df['outcome']               # Output / Dependent F
```

```python
# 6. Split Data 90:10
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size= 0.10, random_state= 123 )
```

```python
X_train, X_test, Y_train, Y_test
```

```python
Lg= LogisticRegression()
```

```python
Lg.fit(X_train,Y_train)
```

```python
Lg.coef_, Lg.intercept_
```

```python
Lg.predict(X_test)
```

```python
# Evaluation
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print("Training Dataset Evaluation : \n")
print("Confusion Matrix : \n")
print(confusion_matrix(Y_train, Lg.predict(X_train) ), " \n")

print("Accuracy Score : ", accuracy_score(Y_train, Lg.predict(X_train)), "\n")

print("Classiification Report : \n ", classification_report(Y_train, Lg.predict(X_train)))
```

```python
print("Training Dataset Evaluation : \n")
print("Confusion Matrix : \n")
print(confusion_matrix(Y_test, Lg.predict(X_test) ), " \n")

print("Accuracy Score : ", accuracy_score(Y_test, Lg.predict(X_test)), "\n")

print("Classiification Report : \n ", classification_report(Y_test, Lg.predict(X_test)))
```
